# Prédécesseurs à partir des successeurs

A partir de la représentation adoptée en cours des graphes comme étant toujours dirigés et donnés sous la forme de liste des successeurs (liste de listes) avec les somments numérotés 0, 1, 2, ..., il s'agit de construir la liste des prédécesseurs. Ceci est fait par la fonction `pred()`ci-dessous.

```python
def pred(g):
    """ on reçoit un graphe donné par les successeurs de chaque sommet,
        et on retourne, avec la même structure, la liste des prédécesseurs
    """
    n = len(g)
    resul = [ [] for _ in range(n) ]
    for x in range(n):
        for y in g[x]:
            resul[y].append(x)
    return resul
```

```python
# Exemple 1 : un triangle dirigé dans le même sens
t = [[1],[2],[0]]
print(pred(t))
```

```python
# Exemple 2 : un chemin du sommet 0 au sommet 4
c = [ [1], [2], [3], [4], [] ]
print(pred(c))
```

Exemple 3 : le graphe de la figure suivante.

```python
g = [ [1,2], [2,3], [1], [2]]
print(pred(g))
```
